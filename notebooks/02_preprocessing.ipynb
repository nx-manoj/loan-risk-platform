## Special Missing Value Handling

The `mths_since_*` features represent the number of months since a specific credit event.

Missing values do not necessarily indicate missing data. In many cases, they mean the event has never occurred.

Since all valid values are greater than or equal to 0, missing values are replaced with -1 as a sentinel value representing "event never occurred."

```python
df[mths_cols] = df[mths_cols].fillna(-1)
```

## Encoding Remaining Categorical Features

The remaining categorical features are encoded into numerical values so that machine learning algorithms can process them.


```python
# Keep only resolved loans
resolved_status = [
    "Fully Paid",
    "Charged Off",
    "Default",
    "Does not meet the credit policy. Status:Charged Off",
    "Does not meet the credit policy. Status:Fully Paid"
]

df = df[df["loan_status"].isin(resolved_status)].copy()
```

```python
df["loan_status"] = df["loan_status"].replace({
    "Fully Paid": 0,
    "Does not meet the credit policy. Status:Fully Paid": 0,

    "Charged Off": 1,
    "Default": 1,
    "Does not meet the credit policy. Status:Charged Off": 1
}).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder


label_encoders = {}

categorical_cols = [
    "home_ownership",
    "verification_status",
    "purpose",
    "addr_state"
]

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```
